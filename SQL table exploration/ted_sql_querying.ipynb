Get a join on our related and fact table for practice

```sql
 select distinct rt.talk_id, tt.title 
 from Talk_fact_dim as tt, Related_talks_id as rt   
    where tt.talk_id = rt.talk_id;
```

Get the titles of the talks with the most related titles

```sql
select tf.talk_id, tf.title, max(rtid.cnt) as related_count
from Talk_fact_dim as tf,  
(select talk_id, count(related_id) 
    as cnt from Related_talks_id group by talk_id) as rtid 
where tf.talk_id = rtid.talk_id group by tf.talk_id, tf.title;

```

Create view for tag counts by name

```sql
create view tag_count as
select tag, count(tag) as tag_count   
from Tag_table 
group by tag;

```

```sql
select * from tag_count order by tag_count desc;
```
